```python
import pandas as pd

subm = pd.read_csv("subm.csv")
subm.head()
```

```python
finals = pd.read_csv("fin.csv",index_col=0)
# Define a function to extract the desired class ID
def get_class_id(row):
    class_id = eval(row['class_id'])[0] if isinstance(row['class_id'], str) else row['class_id'][0]
    return class_id

# Apply the function to each row of the dataframe to extract the desired class ID
finals['new_class_id'] = finals.apply(get_class_id, axis=1)
```

```python
finals[finals.new_class_id=="new_individual"] = -1
finals.drop("class_id",inplace=True,axis=1)
```
